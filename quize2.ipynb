```python
questions = (
    "tall non chalant or short chalant guy?: ",
    "which country won world cup 2006?: ",
    "The best city in Kazakhstan?: ",
    "Who is the best F1 driver?: ",
    "The best modern vocalist?: "
)

options = (
    ("A. tall", "B. short", "C. both"),
    ("A. Spain", "B. Portugal", "C. Italy"),
    ("A. Almaty", "B. Shymkent", "C. Atyrau"),
    ("A. LECLERC", "B. VERSTAPPEN", "C. HAMILTON"),
    ("A. Beyonce", "B. Dimash Kudaibergenuly", "C. Maqan")
)

answers = ("C", "C", "C", "B", "B")

guesses = []
score = 0

for question_num in range(len(questions)):
    print("--------------")
    print(questions[question_num])
    for option in options[question_num]:
        print(option)

    guess = input("Enter (A, B, C): ").upper()
    guesses.append(guess)

    if guess == answers[question_num]:
        print("CORRECT!")
        score += 1
    else:
        print("INCORRECT!")
        print(f"The correct answer was: {answers[question_num]}")

print("---------------------------")
print("          RESULTS          ")
print("---------------------------")

print("Correct Answers: ", end="")
for answer in answers:
    print(answer, end=" ")
print()

print("Your Guesses:    ", end="")
for guess in guesses:
    print(guess, end=" ")
print()

percentage = int((score / len(questions)) * 100)
print(f"Your score: {score}/{len(questions)} ({percentage}%)")
```
